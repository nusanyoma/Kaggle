# 欠損

```python
def missing_table(df):
    null_val = df.isnull().sum()
    percent = 100 * null_val / len(df)
    missing = pd.concat([null_val, percent], axis=1)
    missing_columns = missing.rename(columns={0:"欠損数", 1:"%"})
    return missing_columns

missing_table(train)
```

```python
missing_table(test)
```

# 欠損補完

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")

missing_table(train)
```

```python
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].median())

missing_table(test)
```

# ダミー変数に変換

```python
train_d = pd.get_dummies(train, drop_first=True, columns=["Sex", "Embarked"])
test_d = pd.get_dummies(test, drop_first=True, columns=["Sex", "Embarked"])
train_d
```

```python
test_passenger_ids = test.pop("PassengerId")
train_d.drop(["PassengerId"], axis=1, inplace=True)
train_d
```

# モデル構築

```python
# ランダムフォレスト
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

target = train_d["Survived"].values
features1 = train_d[["Pclass", "Sex_male", "Age", "Fare"]].values

model_RF1 = RandomForestClassifier()
model_RF1.fit(features1, target)

test_features1 = test_d[["Pclass", "Sex_male", "Age", "Fare"]].values

pred = model_RF1.predict(test_features1)
```
