```python
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sub_path = "./outputs/prediction"
all_files = os.listdir(sub_path)
all_files
```

```python
import warnings
warnings.filterwarnings("ignore")
outs = [pd.read_csv(os.path.join(sub_path, f), index_col=0) for f in all_files]
concat_sub = pd.concat(outs, axis=1)
cols = list(map(lambda x: "mol" + str(x), range(len(concat_sub.columns))))
concat_sub.columns = cols
concat_sub.reset_index(inplace=True)
concat_sub.head()
ncol = concat_sub.shape[1]
```

```python
# check correlation
concat_sub.iloc[:,1:ncol].corr()
```

```python
# get the data fields ready for stacking
concat_sub['m_max'] = concat_sub.iloc[:, 1:ncol].max(axis=1)
concat_sub['m_min'] = concat_sub.iloc[:, 1:ncol].min(axis=1)
concat_sub['m_mean'] = concat_sub.iloc[:, 1:ncol].mean(axis=1)
concat_sub['m_median'] = concat_sub.iloc[:, 1:ncol].median(axis=1)
```

```python
cutoff_lo = 0.8
cutoff_hi = 0.2
```

```python
import datetime
today = str(datetime.date.today())
```

```python
concat_sub['isFraud'] = concat_sub['m_mean']
concat_sub[['TransactionID', 'isFraud']].to_csv(f'stack_mean_{today}.csv', 
                                        index=False, float_format='%.6f')
```

```python
concat_sub['isFraud'] = concat_sub['m_median']
concat_sub[['TransactionID', 'isFraud']].to_csv(f'stack_median_{today}.csv', 
                                        index=False, float_format='%.6f')
```
